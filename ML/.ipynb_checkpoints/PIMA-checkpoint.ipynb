```python
#### 머신러닝을 활용한 데이터 분석
# 정확도 정밀도 재현율
from sklearn.metrics import accuracy_score,f1_score,precision_score,recall_score
#ROC AUC 오차행렬
from sklearn.metrics import confusion_matrix, roc_auc_score,roc_curve,precision_recall_curve
# 데이터셋 분류: 트레인테스트셋분류,k-fold,s-kfold, gridsearchcv
from sklearn.model_selection import train_test_split,KFold,StratifiedKFold,GridSearchCV
# 분류모델 테스트
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier

from sklearn.linear_model import LogisticRegression
# 경고 무시 LogisticRegression solver 및 maxiter 문제 
import warnings
warnings.filterwarnings('ignore')
```

```python
X=pidf.iloc[:,:-1] # 숫자기반 split
y=pidf.loc[:,['Outcome']] # 컬럼기반 split
X
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=123,test_size=0.2,stratify=y)
len(X_train)
```

```python
lrm=LogisticRegression(solver='lbfgs', max_iter=1000)
lrm.fit(X_train,y_train)

pred=lrm.predict(X_test)
# 정확도
acc=accuracy_score(pred,y_test)
acc
```

```python
# 모델평가지표 함수
def modelacc(y_test,pred):
    confmat=confusion_matrix(y_test,pred)
    #print('오차행렬\n-------\n',confmat)
    p_score=precision_score(y_test,pred)
    pscore=np.round(p_score,4)
    print('정밀도:',pscore, end='')
    r_score=recall_score(y_test,pred)
    rscore=np.round(r_score,4)
    print('재현율:',rscore, end='')
    f1_s=f1_score(y_test,pred)
    f1score=np.round(f1_s,4)
    print('f1 스코어:',f1score)
    return([pscore,rscore,f1score])
modelacc(y_test,pred)
```

```python
ppt=lrm.predict_proba(X_test)
print(' 입력형태:',X_test.shape,'\n 출력형태:',ppt.shape)
ppt[:3]
pptdf=pd.DataFrame(ppt)
pptdf['result']=pred
pptdf.head()
```

```python
from sklearn.preprocessing import Binarizer
a=[0.3,0.5,0.7]
aa=np.array(a).reshape(-1,1)
b=Binarizer(threshold=0.5).fit_transform(aa)
print(b)
```

```python
# 임계치 기반의 재현율 정밀도 f1스코어 반환함수
def modelthr(ppt,thre=0.5):
    ppt1=ppt[:,1].reshape(-1,1)
    #ppt1
    pptres=Binarizer(threshold=thre).fit_transform(ppt1)
    #pptres
    res=modelacc(pptres,y_test)
    return res
```

```python
tstep=np.arange(0.1,1.0,0.05)
tstep
```

```python
accres=[]
for step in tstep:
    accs=modelthr(ppt,step)
    accres.append(accs)
accres
```

```python
# 분류 결정 임계값
accdf=pd.DataFrame(accres,columns=['Precision','Recall','F1'])
accdf.index=tstep
accdf.head()
```

```python
#정밀도 재현율 F1 스코어 그래프
accdf.plot()
```

```python
# 쉽게 해주는 라이브러리
from sklearn.metrics import precision_recall_curve
ppt1=ppt[:,1]
precision,recalls,thresholds=precision_recall_curve(y_test,ppt1)
print(ppt1.shape)
print(precision.shape)
print(recalls.shape)
print(thresholds.shape)
idx=np.arange(1,thresholds.shape[0],10)
thre=np.round(thresholds[idx],2)
prec=np.round(precision[idx],2)
reca=np.round(recalls[idx],2)
print(len(idx),len(thre),len(prec),len(reca))
data={'precision':prec,'recalls':reca}
prcdf=pd.DataFrame(data)
prcdf.index=thre
prcdf.plot()
```

### ROC Curve
* 입력파라미터
    - y_true: 실제 클래스 값
    - y_score: predict_prob()에서 반환된 값의 1에 해당하는 컬럼
* 출력
    - FPR: FP/(FP+TN)
    - TPR: TPR 변화값
    - thresholds

```python
from sklearn.metrics import roc_curve,roc_auc_score
fpr,tpr,thresholds=roc_curve(y_test,ppt1)
print(ppt1.shape)
print(fpr.shape)
print(tpr.shape)
print(thresholds.shape)
idx=np.arange(1,thresholds.shape[0],10)
thre=np.round(thresholds[idx],2)
fpr=np.round(fpr[idx],2)
tpr=np.round(tpr[idx],2)
print(len(idx),len(thre),len(prec),len(reca))
data={'TPR':tpr,'FRP':fpr}
rocdf=pd.DataFrame(data)
```

```python
auc_score=roc_auc_score(y_test,ppt1)
auc_s=np.round(auc_score,3)
print(auc_s)
plt.plot(fpr,tpr)
plt.xlabel('FPR(1-Sensitivity)')
plt.ylabel('TPR(Recall)')
plt.title('ROC Curve( AUC:'+str(auc_s)+')')
plt.show()
```
