```python
# np arange로 범위 생성
x = np.arange(-4.5, 5, 0.5)
# y값 (2차 함수) 생성
y = 2 * x ** 2
plt.plot(x, y)
# 제목 쓰기
plt.title('graph1')
# 최종 그래프 출력
plt.show()
```

```python
# np arange로 범위 생성
x = np.arange(-4.5, 5, 0.5)
# y1, 2, 3값 생성
y1 = 2 * x ** 2
y2 = 5 * x + 30
y3 = 4 * x ** 2 + 10
# 라벨 지정
plt.plot(x, y1, label = 'Y1')
plt.plot(x, y2, label = 'Y2')
plt.plot(x, y3, label = 'Y3')  # plt.plot(x, y1, x, y2, x, y3)도 같은 값으로 출력
# 지정 라벨의 범례 표시
plt.legend()
plt.title('function')
plt.show()
```

```python
plt.plot(x, y1)
# 새로운 그래프 창 생성
plt.figure()
plt.plot(x, y2)
plt.show()
```

```python
# np arange로 범위 생성
x = np.arange(-4.5, 5, 0.5)
# y1, 2, 3값 생성
y1 = 2 * x ** 2
y2 = 5 * x + 30
y3 = 4 * x ** 2 + 10
# 라벨 지정
# figure(n) n지정으로 특정 그래프에 겹쳐 그리기
plt.figure(1)
plt.plot(x, y1, label = 'Y1')
plt.title('graph1')
plt.legend()

plt.figure(2)
plt.plot(x, y2, label = 'Y2')
plt.title('graph2')
plt.legend()

plt.figure(1)
plt.plot(x, y3, label = 'Y3')
plt.legend()
plt.show()
```

```python
# np arange로 범위 생성
x = np.arange(0, 10, 0.1)
# y1, 2, 3, 4값 생성
y1 = 0.3 * (x - 5) ** 2 + 1
y2 = -1.5 * x + 3
y3 = np.sin(x) ** 2 
y4 = 10 * np.exp(-x) + 1
# 라벨 지정
ax0 = plt.subplot(2, 2, 1)
plt.plot(x, y1, label = 'Y1')
ax1 = plt.subplot(2, 2, 2)
plt.plot(x, y2, label = 'Y2')
ax2 = plt.subplot(2, 2, 3)
plt.plot(x, y3, label = 'Y3')
ax3 = plt.subplot(2, 2, 4)
plt.plot(x, y4, label = 'Y4')  # plt.plot(x, y1, x, y2, x, y3)도 같은 값으로 출력
plt.show()
```

```python
# 초기화 
plt.figure()
# 구역 정리
ax1 = plt.subplot(2, 2, 1)
ax2 = plt.subplot(2, 2, 2)
ax3 = plt.subplot(2, 2, 3)
ax4 = plt.subplot(2, 2, 4)
# ax별로 그래프 그리기
ax1.plot(x, y1)
ax1.set_title('graph1')  # ax기반은 명령어 대부분 set_~ 시작
ax2.plot(x, y2)
ax2.set_title('graph2')
ax3.plot(x, y3)
ax3.set_title('graph3')
ax4.plot(x, y4)
ax4.set_title('graph4')
```

```python
# 한 번에 구조 만들기
fig, ax = plt.subplots(2, 2)
# 그래프 간 간격 조정
fig.subplots_adjust(wspace = 0.5, hspace = 0.5)
# ax별로 그래프 그리기
ax[0, 0].plot(x, y1)
ax[0, 1].plot(x, y2)
ax[1, 0].plot(x, y3)
ax[1, 1].plot(x, y4)
# ax별 그래프 Title
ax[0, 0].set_title('figure1')
ax[0, 1].set_title('figure2')
ax[1, 0].set_title('figure3')
ax[1, 1].set_title('figure4')
# 기타 옵션 블럭
ax[0, 0].set_xticks([])
ax[0, 1].set_xticks([])
plt.show()
```

```python
# np arange로 범위 생성
x = np.arange(0, 10, 0.1)
# y1, 2, 3, 4값 생성
y1 = 0.3 * (x - 5) ** 2 + 1
y2 = np.sin(x) ** 2 
# 라벨 지정
ax1 = plt.subplot(2, 1, 1)
plt.plot(x, y1, label = 'Y1')
# 축 공유가 되었으니 중복은 숨김
plt.xticks(visible = False)
plt.subplot(2, 1, 2, sharex = ax1)
plt.plot(x, y2, label = 'Y2')
plt.show()
```

```python
# 1, 2 1행 2열 그래프 표시
# y1, 2값 생성
y1 = 0.3 * (x - 5) ** 2 + 1
y2 = -1.5 * x + 3
# 라벨 지정
ax1 = plt.subplot(1, 2, 1)
plt.plot(x, y1, label = 'Y1')
plt.title('graph1')
# 축 공유가 되었으니 중복은 숨김
ax2 = plt.subplot(1, 2, 2, sharex = ax1)
plt.plot(x, y2, label = 'Y2')
plt.title('graph2')
plt.yticks(visible = False)
plt.show()
```

```python
x = np.linspace(-4, 4, 100)
y1 = x ** 3
y2 =  10 * x ** 2 - 2

plt.figure(1)
plt.plot(x, y1, label = 'Y1')
plt.plot(x, y2, label = 'Y2')
plt.legend()
plt.show()
# 최소값 최대값 정해서 확대해서 보기 (겹치는 부분이 잘 안보이는 경우)
plt.figure(2)
plt.plot(x, y1, label = 'Y1')
plt.plot(x, y2, label = 'Y2')
plt.xlim(-1, 1)  # xmin ~ xmax
plt.ylim(-3, 3)  # ymin ~ ymax
plt.legend()
plt.show()
```

```python
# 가로 세로 다른 개수로 서브플롯 만들기
# 구조 만들기
ax1 = plt.subplot(212)
ax2 = plt.subplot(221)
ax3 = plt.subplot(222)
# 그래프 그리기
ax1.plot(x, y1)
ax2.plot(x, y2)
ax3.plot(x, y3)
# 옵션주기
ax1.margins(10)  # 줌 아웃 테두리 과다하게 주기
ax2.margins(-0.1, -0.1)  # 줌 들어가기
ax3.margins(0.5, 10)  # 가로세로 다른 비율로 줌 주기
```

```python
# subplots 뺄 때 ax 개별 지정
# gridspec_kw 간격 벌릴 때
# figsize = () 그래프 사이즈 조절
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows = 2,
                                             ncols = 2,
                                             figsize = (5, 10))
fig.subplots_adjust(hspace = 0.5)
# 그래프 블럭
ax1.plot(x, y1)
ax2.plot(x, y2)
ax3.plot(x, y3)
ax4.plot(x, y4)
# 옵션 블럭
ax1.set_title('graph1', fontsize = 15)
ax1.set_xlabel('x axis', fontsize = 15)
ax1.set_ylabel('y axis', fontsize = 15)
```

## 그래프 꾸미기
### 출력 형식 지정
#### fmt = '[color][line_style][marker]'
|컬러 약어|  컬러  | 라인 스타일 |선 스타일         | 마커 약어| 마커                   |
|:-------:|:------:|:-----------:|:---------------:|:--------:|:----------------------:|
|b        |  파랑  | -           | 실선            |o          |원 모양                 |
|g        |  초록  | --          | 파선            |^, v, < ,> |삼각형 위, 아래, 왼, 오 |
|r        |  빨강  | :           | 점선            |s          |사각형                  |
|c        |  청녹  | -.          | 파선 점선 혼합선|p          |오각형                  |
|m        |  자홍  |             |                 |h, H       |육각형                  |
|y        |  노랑  |             |                 |*          |별                      |
|k        |  검정  |             |                 |+          |더허기                  |
|w        |  하양  |             |                 |x, X       |x, 채워진X              |
|w        |  하양  |             |                 |D, d       |다이아, 얇은 다이아     |

```python
x = np.arange(-3, 3, 0.1)
len(x)

plt.figure(1)
y0 = np.ones(60)  # 상수값
print('상수(y0):', y0[:10])
plt.plot(x, y0, '--',label = 'x^0')

y1 = x * 1.5  # 일차함수
plt.plot(x, y1, 'r-',label = 'x^1')

y_1 = x * -1.5  # 일차함수 우하향
plt.plot(x, y_1, 'g--',label = '-x^1')

y11 = x * 1.5 + 1  # 일차함수 우상향
plt.plot(x, y11, 'b-.',label = 'x1^+')

y1_1 = x * 1.5 - 1  # 일차함수 우상향
plt.plot(x, y1_1, 'y:',label = 'x^1-')

plt.title('first-order function')
plt.legend()
plt.show()
######################################
# 2차 함수
plt.figure(2)
y2 = x ** 2
plt.plot(x, y2, 'r-',label = 'x^2')

y42 = 4 * x ** 2  # 좁은 곡선
plt.plot(x, y42, 'g-',label = '4x^2')

y_42 = -4 * x ** 2  # 좁은 곡선 (기울기 반대)
plt.plot(x, y_42, 'b-',label = '-4x^2')

y24 = x ** 2 + 4  # 좁은 곡선 이동
plt.plot(x, y24, 'y-',label = 'x^2+4')

plt.title('quadratic function')
plt.legend()
plt.show()
```

```python
# 곡선 연습
x = np.arange(-3, 3, 0.1)
len(x)
y = x ** 2
# 관측치 강제 생성
z = [3 * yi + np.random.rand(1)[0]+ np.random.rand(1)[0] * 10 + 3 for yi in y]
# 가중치와 편차 수작업으로 찾아내기
# 직관적으로 만들어낸 설명 선
w = 2.5 * x ** 2 + 10
plt.plot(x, w)
plt.plot(x, z, 'r+')
plt.show()
```

```python
markers = ['o', '^', '>', '<', 'v', 's', 'p', '*', 'd']
linestyle = [':', '-', '--', '-.']
colors = 'bgrcmyk'
i = 0
for m in markers:
    i += 1
    x = [i, i + 1]
    y = [i * 2, (i + 1) * 2]
    linei = i % 4
    colori = i % 7
    plt.plot(x, y, 'g' + linestyle[linei])
    plt.plot(i, i * 2, colors[colori] + m, label = colors[colori] + ' ' + m)
    plt.legend(loc = 'lower right')  # 범례 표시 + 위치
    plt.grid(True)  # 그리드 그리기
    plt.title('fmt options')  # 제목 표시
```
