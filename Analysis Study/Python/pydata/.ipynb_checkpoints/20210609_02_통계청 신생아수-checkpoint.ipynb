```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib as mp
```

```python
fm = mp.font_manager.FontManager()
plt.rcParams['font.family'] = 'Malgun Gothic'
```

```python
df_kosis = pd.read_csv('pdsample/population_kosis_1997_2019.csv',encoding='cp949')
```

```python
#melt는 데이터를 재구조화 시켜줄 때 사용한다.
#칼럼들이 variable이라는 열로 만들어지고
#컬럼의 값들이 value라는 열로 들어가게됨

#열들을 하나의 열로 만듬 -> 열에 적합한 값을 하나의 열로 또 만듬.

df = df_kosis.melt(id_vars='시군구별')
df.head()
```

```python
df.shape
```

```python
df['시군구별'].unique()
```

```python
len(df['시군구별'].unique())
```

```python
df[df['시군구별']=='시군구별']
```

```python
df = df[df['시군구별']!='시군구별'].copy()
```

```python
df
```

```python
df.head()
```

```python
df['variable']
type(df['variable'])
print(type(df['variable'].str))
```

```python
type(df['variable'].str.split(',',expand = True))
```

```python
df['연도'] = df['variable'].str.split('.', expand=True)[0]
df['월'] = df['variable'].str.split('.', expand=True)[1]
df['성별'] = df['variable'].str.split('.', expand=True)[2]
```

```python
df.head()
```

```python
df.tail()
```

```python
df
```

```python
df.info()
```

```python
df['성별'].unique()
```

```python
df['성별'].nunique()
```

```python
df['성별'].isnull().sum()
```

```python
df['성별'] = df['성별'].fillna('전체')
```

```python
df['성별'].nunique()
```

```python
df['성별'] = df['성별'].replace('1','남').replace('2','여')
```

```python
df
```

```python
df['성별'].value_counts()
```

```python
df = df.rename(columns = {'variable':'기간','value':'출생아수'})
```

```python
df
```

```python
df.sample(10)
```

```python
df['출생아수'] = df['출생아수'].replace('-',np.nan) 
```

```python
df['출생아수'].isnull().sum()
```

```python
df.info()
```

```python
df['출생아수'] = df['출생아수'].replace(-1,np.nan)
```

```python
df['출생아수'] = df['출생아수'].astype(float)
```

```python
df.info()
```

```python
df['출생아수'].isnull().sum()
```

```python
df.loc[8616]
```

```python
df['출생아수'].describe()
```

```python
df_all = df[(df['시군구별'] == '전국') & (df['성별']=='전체')]
```
