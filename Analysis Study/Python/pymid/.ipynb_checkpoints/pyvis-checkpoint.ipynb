```python
# 한장에 그리기
# np arange 로 범위 생성
x=np.arange(-2,2,0.1)
# y 값 생성
y=x**2
z=x*2+3
w=x*3+1
# 라벨지정 여러 그래프 그리기
plt.plot(x,y,label='x')
plt.plot(x,z,label='z')
plt.plot(x,w,label='w')
# 지정 라벨의 범례 표시
plt.legend()
# 제목쓰기
plt.title('graph1')
# 최종그래프 출력하기
plt.show()
```

```python
#figure(n) n 이 없는 경우 새창 생성
plt.figure()
plt.plot(x,y,label='x')
plt.legend()
plt.figure()
plt.plot(x,z,label='z')
plt.figure()
plt.plot(x,w,label='w')
```

```python
#figure(n) n 지정으로 특정 그래프에 겹쳐 그리기
plt.figure(1)
plt.plot(x,y,label='x')
plt.title('graph1')
plt.legend()

plt.figure(2)
plt.plot(x,z,label='z')

plt.figure(1)
plt.plot(x,w,label='w')
plt.title('graph2')
plt.show()
```

```python
# subplot(m,n,p) m*n 그래프 생성후 p번째 그리기
# figure 와 유사한데 구조적 그래프 생성
```

```python
x=np.arange(-2,2,0.1)
# y 값 생성
y=x**2
z=x*2+3
v=np.arange(-2,1,0.1)
w=v*3+1
#2,1 2행 1열 그래프 표시
plt.subplot(2,1,1)
plt.plot(x,y)
plt.plot(x,z)
plt.title('graph1')
plt.subplot(2,1,2)
plt.plot(v,w)
plt.show()
```

```python
#2,1 2행 1열 그래프 표시
ax1=plt.subplot(2,1,1)
plt.plot(x,y)
plt.plot(x,z)
plt.title('graph1')
# 축 공유가 되었으니 중복은 숨김
plt.xticks(visible=False)
# 축 공유를 통한 비교의 대상을 표준화
ax2=plt.subplot(2,1,2,sharex=ax1)
plt.plot(v,w)
plt.title('graph2')
plt.show()
```

```python
#1,2 1행 2열 그래프 표시
ax1=plt.subplot(1,2,1)
plt.plot(x,y)
plt.plot(x,z)
plt.title('graph1')
# 축 공유를 통한 비교의 대상을 표준화
ax2=plt.subplot(1,2,2,sharex=ax1)
plt.plot(v,w)
# 축 공유가 되었으니 중복은 숨김
plt.yticks(visible=False)
plt.title('graph2')
plt.show()
```

```python
#2,2 2행 2열 그래프 표시
# 복잡해지면 구조잡고 그래프그리기
# 그래프 초기화
plt.figure()
ax0=plt.subplot(2,2,1)
plt.plot(x,y)
plt.title('graph0')

ax1=plt.subplot(2,2,2)
plt.plot(x,z)
plt.title('graph1')
# 축 공유를 통한 비교의 대상을 표준화
ax2=plt.subplot(2,2,4)
plt.plot(v,w)
# 축 공유가 되었으니 중복은 숨김
plt.yticks(visible=False)
plt.title('graph2')
plt.show()
```

```python
#초기화
plt.figure()
#구역정리
ax1=plt.subplot(2,2,1)
ax2=plt.subplot(2,2,2)
#ax3=plt.subplot(2,2,3)
ax4=plt.subplot(2,2,4,sharex=ax2)

#ax 별로 그림 그리기
ax1.plot(x,y)
ax1.set_title('graph1')

ax2.plot(x,z)
ax2.set_title('graph2') # ax기반은 대부분 set_~ 시작
ax2.set_xticks([]) # ax 기반의 set_~ 옵션이 상이함

ax4.plot(x,y)
ax4.set_title('graph4')


plt.show()
```

```python
# 한번에 구조 만들기 s 에 주의
fig,ax=plt.subplots(2,2)
# 배열기반의 축지정
# 그래프 블럭
ax[0,0].plot(x,y)
ax[1,0].plot(x,y)
ax[0,1].plot(x,y)
ax[1,1].plot(x,y)
#타이틀 블럭
ax[0,0].set_title('graph1')
ax[1,0].set_title('graph2')
ax[0,1].set_title('graph3')
ax[1,1].set_title('graph4')
# 기타 옵션 블럭
ax[0,0].set_xticks([])
ax[0,1].set_xticks([])
```

```python
ax1=plt.subplot(1,2,2)
ax3=plt.subplot(2,2,2)
ax2=plt.subplot(1,2,1)
# 그래프 그리기
ax1.plot(x,y)
ax2.plot(x,y)
ax3.plot(x,y)
# 옵션주기
ax1.margins(10) # 줌 아웃 테두리 과다하게 주기
ax2.margins(-0.1,-0.1) # 줌들어가기
ax3.margins(0.5,10)# 가로세로 다른 비율로 줌주기
```

```python
# 가로 세로 다른 갯수로 서브플롯 만들기
# 구조 만들기
ax1=plt.subplot(212)
ax2=plt.subplot(221)
ax3=plt.subplot(222)
# 그래프 그리기
ax1.plot(x,y)
ax2.plot(x,y)
ax3.plot(x,y)
# 옵션주기
ax1.margins(10) # 줌 아웃 테두리 과다하게 주기
ax2.margins(-0.1,-0.1) # 줌들어가기
ax3.margins(0.5,10)# 가로세로 다른 비율로 줌주기
```

```python
# subplots 뺄때 ax 개별지정
# gridspec_kw 간격 벌릴때
# figsize=(가로,세로)로 그림 비율 조정
fig,((ax1,ax2),(ax3,ax4))=plt.subplots(nrows=2,
                                       ncols=2,
                                       figsize=(5,10),
                                       gridspec_kw=dict(left=0.1, right=0.9,
                                        bottom=0.1, top=0.9),
                                      )
# 그래프 블럭
ax1.plot(x,y)
ax2.plot(x,y)
ax3.plot(x,y)
ax4.plot(x,y)
# 옵션블럭

ax1.set_title('graph1',fontsize=20)
ax1.set_xlabel('x axis',fontsize=20)
ax1.set_ylabel('y axis',fontsize=20)
```
