### 자연어 처리(NLP: Natural Language Processing)
* 목적
    - 언어-> 디지털화 프로세싱
    - 텍스트의 의미 성격 등을 과학적으로 분석하기위함
* 차이
    - 상대적 라벨링: 정형화(객관화) 데이터
    - 객관적 숨은 의미 언어간의 거리
* 방법
    - 글의 요소
        * copus 글덩어리(코퍼스:책)
        * paper 문서(페이지:의미 작음)
        * paragraph 문단(주제의단위: 의미 좀 작음)
        * sentence 문장(주어 서술어 목적어등을 통해 의미 확보)
        * word 어휘 + 문장부호
* 분야
    - 텍스트 분류(Text Classification)
        - 주제분류(가십, 연예, 정치)
        - 스팸메일 분류
    - 감성분석(Sentiment Analysis)
        - 긍부정 판단
        - 여론조사 의견강약
    - 텍스트 요약(Summarization)
        - 주요주제나 중심사상 발췌
    - 유사도 측정
        - 해당하는 말이 같은 주제를 말하고 있는가
        - 표절과 유사
* NLP 핵심 요소
    - 형태소
        - 정의: 의미가 있는 가장 작은말 더 분석하면 뜻이 없어지는 말의 단위
        - 종류
            - 문법(형식) 형태소
                - 어휘형태소 간의 관계나 기능을 나타내는 형태소
            - 어휘(실질) 형태소
                - 어휘적 의미가 있는 형태소( 상태, 동작)
                - 명사, 동사, 형용사, 부사
    - 텍스트 분석 프로세스
        * 프로세스
            - 클렌징: 대소문자 변환 특수문자 제거
            - 불용어 제거 (Stop word) 1글자 문장부호
            - 토큰화(형태소 분리)
                - 어간추출(Stemming)
                - 표제어 추출(Lematization)
        * 기초 프로세싱
            - BOW(Bag of Words) 
                - 토큰화된 단어를 단어별 빈도등으로 구별하여 분석

```python
file='src/갈매기.txt'
with open(file,encoding='utf-8') as fn:
    # 클렌징
    res=[]
    for line in fn.readlines():
        lstrip=line.replace('\ufeff','').strip()
        if(lstrip):
            res.append(lstrip)
print(len(res),res[:3])
```

```python
wordbook={}
def countWord(senten):
    words=senten.split(' ')
    for word in words:
        if(len(word)>1):
            if(word in wordbook.keys()):
                wordbook[word]+=1
            else:
                wordbook[word]=1
    return wordbook
#countWord(res[1])
```

```python
wordbook={}
for r in res:
    countWord(r)
len(wordbook)
```
