### 수익률
* 계산
    - (매도금액- 매입금액+ 현금흐름)/매입금액
*예    
    * A가 사기 맞았을때
    * A :  (매도금액- 매입금액+ 현금흐름)/매입금액=(0-100+120)/100
        - A의 수익률 20%
    * B의 수익률
        - (0-100+40)/100=-60% 손실
    
* 수익에는 리스크가 반드시 따른다
    - 기대수익률 = 무위험 수익률+리스크 프리미엄

### 기업가치 분석
* 기업 평가  지표
    * 기업의 재무및 경영상태를 보다 객관적으로 알기위한 지표
    * 재무제표, 손익계산서, 현금흐름표
* 재무비율 지표
    - 레버리지 비율(부채비율)
        - (총부채/자기자본)*100
        -  200% 안쪽을 권고
    - 이자 보상 비: 갚을수 있는 능력
        - 영업이익/이자비용
        - 1 보다 작으면-> 일해서 이자도 못갚는다.
    - 상환능력이 있는가: 유동비율
        - 유동비율 =(유동자산/유동부채(단기부채))*100
        - 단기부채 상환능력 부족시: 흑자도산
        - 당좌비율 = (유동자산-재고자산)/유동부채*100 
    - 활동성 지표: 액티브한가
        -  기업이 보유하고 있는 자산을 얼마나 잘 활용하고 있는가
        - 장치산업은 상대적으로 낮은 활동성 비율(동종업계비교)
            - 자산회전비 = 매출액 / 총자산
            - 재고자산 회전비= 매출액 / 재고자산
            - 평균회수기간=365일/(매출액/매출채권)
    - 수익성 지표
        - 총자산 이익률(ROA:Return on Asset) 차입자본 포함
            - ROA=(순이익/ 총자산)*100
        - 자기자본 이익률(ROE: Return on Equity)
            - ROE=(순이익/ 자기자본)*100
        - 년도별 추적을 통해 -> 기업가치 변화 -> 투자가치 판단
    - 한계를 알고 분석의 위험요소를 회피:여러 회계 요인을 통합 분석

        
        
        
        

### 주가 배수 평가
- 투자를 위한 최적의 판단 지표
    *  PER(Price Earning Ratio):주가를 1주당 이익으로 나눔
        - 주가 / 주당순이익(EPS) => 미래 가치 포함 /현재
        - 동종업계및 시계열 분석  
    *  PBR(Price Bookvalue Ratio)주가 장부 가치비율
        - 주가/주당순자산(BPS)
        - 청산가치 비율
    * EV/EBITA 원금 회수기간
        - EV (Enterpirse Value) 기업가치= 시가총액(자기자본)+순차입금
        - EBITA(Earning Before Interest,Tax,Depreciation & Amortization)
            - 이자 세금 감가상각비 차감하기 이전 이익
            - 실질 현금흐름에 노이즈를 없애기 위해
