```python
import requests
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import sqlite3

from selenium import webdriver
#98.0.4758.82

from bs4 import BeautifulSoup as bsp
```

```python
def getreq(url='',tag='body'):
    h = {'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/63.0.3239.132 Safari/537.36'}
    src = requests.get(url, headers = h).text.strip()
    html=bsp(src,'html.parser')
    tgts=html.select(tag)
    return tgts
```

```python
url='https://movie.naver.com/movie/sdb/rank/rmovie.naver?sel=pnt&date=20220206&page=1'

tgts=getreq(url,'table.list_ranking tbody tr')
rooturl='https://movie.naver.com'
for tgt in tgts[:10]:
    tds=tgt.select('td')
    if(len(tds)>1):
        title=tds[1].text.strip()
        hurl=tds[1].select('a')[0]['href']
        turl=rooturl+hurl
        print(title,rooturl+hurl)
        mtgt=getreq(turl,'div.article')
        ## title
        titlesrc=mtgt[0].select('div.mv_info .h_movie2')[0].text
        tsplit=titlesrc.split(',')
        tcomma=len(tsplit)
        ####### title
        if tcomma==1:
            etitle=''
            byear=tsplit[0].strip()
        elif tcomma==2:
            etitle=tsplit[0].strip()
            byear=tsplit[1].strip()
        elif tcomma==3:
            etitle=tsplit[1].strip()
            byear=tsplit[2].strip()
        ######### score
        
        scores=mtgt[0].select('div.main_score div.star_score')
        ss=scores[0].text.strip()
        if(ss.find('람객 평점 없음')>-1):
            auds=0
            spcs=0
            ntzs=float(scores[1].text.strip())
        else:
            s2=scores[2].text.strip()
            if(s2.find('0점')>-1):
                auds=float(ss[-4:])
                spcs=0
                ntzs=float(scores[1].text.strip())
            else:
                auds=float(ss[-4:])
                spcs=float(scores[1].text.strip())
                ntzs=s2
        #print('score:',len(scores),':',auds,spcs,ntzs)
        
        #### infosepc
        specs=mtgt[0].select('dl.info_spec dd')
        spans=specs[0].select('span')
        for s in spans:
            print(s.text.strip())

        #story=mtgt[0].select('div.story_area')[0].text.replace('줄거리','').strip()
        
        #print(etitle,byear,auds,spcs,ntzs)
```
